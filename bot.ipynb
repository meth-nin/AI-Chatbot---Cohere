```python
import cohere
import os
co = cohere.ClientV2(os.getenv("COHERE_API_KEY"))
```

```python
message = input("Enter your message: ")

system_message = "You are a helpful assistant that provides information about movies. Only give a brief summary of the movie, imdb ratings of the movie asked about, and the release year of the movie. Do not provide any additional information. Only answer questions about movies and nothing else"

messages = [{'role': 'system', 'content': system_message}, 
            {'role': 'user', 'content': message}]

response = co.chat(
    model= "command-a-03-2025",
    messages= messages
)

print(response.message.content[0].text)
```
