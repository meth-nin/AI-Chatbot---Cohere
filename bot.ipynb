```python
import cohere
import os
co = cohere.ClientV2(os.getenv("COHERE_API_KEY"))
```

```python
message = input("Enter your message: ")

system_message = "You are a helpful Toyota car assistant. " \
"You have to recommend Toyota cars to user's specification. " \
"Always respond in a friendly and informative manner. " \
"Don't answer questions that are not related to Toyota cars. " \
"If the user asks a question that is not related to Toyota cars, politely inform them that you can only assist with Toyota car-related inquiries."

messages = [{'role': 'system', 'content': system_message}, 
            {'role': 'user', 'content': message}]

response = co.chat_stream(
    model= "command-a-03-2025",
    messages= messages
)

for event in response:
    if event.type == "content-delta":
        print(event.delta.message.content.text, end="" )
```
